```python
import matplotlib
import matplotlib.pyplot as plt
import pylab
```

```python
x = [100, 1000, 10000, 100000]
a = [0.036201, 0.015349, 0.002401, 0.0009578326859147506]
b = [0.049311, 0.028033, 0.006167, 0.0026622144609346966]
```

```python
SMALL_SIZE = 8
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
fig = plt.figure()
ax = plt.subplot(1,1,1)


ax.plot(x, a, color='blue', lw=3)

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('niter')
ax.set_ylabel('optimality gap')
fig.savefig('1.pdf', bbox_inches='tight')
```

```python
fig = plt.figure()

ax = plt.subplot(1,1,1)
ax.plot(x, b, color='red', lw=3)

ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('niter')
ax.set_ylabel('standard deviation')

pylab.show()

fig.savefig('2.pdf', bbox_inches='tight')
```
